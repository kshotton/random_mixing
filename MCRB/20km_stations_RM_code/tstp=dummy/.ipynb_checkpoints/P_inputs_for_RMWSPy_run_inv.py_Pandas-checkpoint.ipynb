### Hay Meadow

```python
# Use pd.read_csv to read csv file
HM_Pinputs_df = pd.read_csv(r"C:\Users\b1043453\OneDrive - Newcastle University\OnePlanet PhD\Hydrological_Modelling\Marmot_Creek_Research_Basin\Version 02\Meteorology\Recent_modelling_data\time_series_inputs\HMdailyprecip_nodates.csv", header=None, dtype=float)

# transpose the dataframe
HM_Pinputs_df = HM_Pinputs_df.T

HM_Pinputs_df
```
