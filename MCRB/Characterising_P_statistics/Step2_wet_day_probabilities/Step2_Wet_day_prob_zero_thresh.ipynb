### Plot histogram and probability distributions for wet days at each station in descending order of elevation

### Fisera Ridge 2325 m, 2330 wet days

```python
FRP_df.FR_p_mm.plot(kind='hist', bins=50)
plt.title('Fisera Ridge 2325 m (2005-2016)')
```

```python
# fit distributions
x = np.linspace(1, max(FRP_df.FR_p_mm), 100)
y = FRP_df.FR_p_mm

FRP_df.FR_p_mm.plot(kind='hist', bins=50, density=True, color='lightgrey', label='precipitation histogram')

param = scipy.stats.gamma.fit(y)
print("gamma:", param)
pdf_fitted = scipy.stats.gamma.pdf(x, a=0.76, scale=5.25) # a is shape param
# I tested a range of a from 0.1 to 1, and scale from 1 to 10 and came back to 
# Liz's values of 0.76 and 5.25 as they give a good fit
plt.plot(x, pdf_fitted, label='fitted:gamma')

param = scipy.stats.expon.fit(y)
print("exponential:", param)
pdf_fitted = scipy.stats.expon.pdf(x, param[0], param[1])
plt.plot(x, pdf_fitted, label='fitted:exponential')
plt.xlabel('rainfall (mm)')
plt.xlim(0, 40)
plt.ylim(0, 0.3)
plt.legend()
plt.title('Fisera Ridge 2325 m (2005-2016)')
# plt.show()
plt.savefig(r'C:\Users\b1043453\OneDrive - Newcastle University\OnePlanet PhD\Random_mixing\RMWSPy_Horning_and_Haese_2021\RMWSPy-master\MCRB_examples\MCRB_gauges_only\Characterising_P_statistics\probdist_plots_2005-2016_stns\FRprob_0mm_2005-16.png')
```
